```python
import pandas as pd 
from splinter import Browser
from bs4 import BeautifulSoup
import requests
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://reelgood.com/movies/source/netflix?filter-sort=2'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for x in range(10):

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    titles = soup.find_all('table', class_='css-1179hly')

    for title in titles:
        print('page:', x, '-------------')
        print(title.text)

    browser.find_by_css('button.css-qbp45y').click()
```

```python
title_df = pd.read_html(str(title))
netflix = title_df[0]
netflix
```

```python
netflix_df = netflix.drop(columns=["Unnamed: 0", "Unnamed: 2", "Rating", "Unnamed: 6", "Available On", "Unnamed: 8", "Unnamed: 9"])
netflix_clean = netflix_df.rename(columns= {"Unnamed: 5" : "IMDB Rating"})
netflix_clean
```
